Q1 : Using the following link of Times of India headlines page, find the top 10 headlines and link to read the entire news corresponding to the heading.

https://timesofindia.indiatimes.com/home/headlines

Creating DataFrame

```python
df = pd.DataFrame(columns=['Headline', 'Link'])
```

Function for Scraping

```python
def get_headlines():
    url = 'https://timesofindia.indiatimes.com/home/headlines'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for d in soup.find_all('span', attrs={'class': "w_tle"})[:10] :
        Title = d.find('a').text
        Link = d.find('a')['href']
        data=[]
        if Title is not None:
            data.append(Title)
        else:
            data.append(' ')
        if Link is not None:
            data.append('https://timesofindia.indiatimes.com' + Link)
        else:
            data.append(' ')
        df.loc[len(df)]=data
    return df
```

Printing the data

```python
get_headlines()
```

Q2: From Linkedin website, extract job title, job company, job link for the jobs that appears on the first page when you search for a particular profile in a particular location. For example, your search may be for software engineer in Delhi. For this search store job title, job link, and company name that appears on the first page of your search

Creating DataFrame

```python
fd = pd.DataFrame(columns=['Job_Title', 'Job_Company', 'Job_Link'])
```

Function for Scraping

```python
def getData(job, place):
    url = f'https://in.linkedin.com/jobs/search?keywords={job}&location={place}&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0&original_referer='
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    fd = pd.DataFrame(columns=["Job Title", "Company", "Link"])

    for d in soup.find_all('div', attrs={'class': "base-search-card__info"}):
        title = d.find('h3', attrs={'class': "base-search-card__title"})
        company = d.find('h4', attrs={'class': 'base-search-card__subtitle'})
        imp=d.find_parent()
        link=imp.find('a',attrs={'class':"base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]"})
        
        data = []
        if title is not None:
            data.append(title.text.strip())
        else:
            data.append('')
        if company is not None:
            data.append(company.text.strip())
        else:
            data.append('')
        if link is not None:
            data.append(link['href'])
        else:
            data.append('')
        
        fd.loc[len(fd)] = data

    return fd
```

Getting Data for Software Engineer in Delhi

```python
getData('Software Engineer', 'Delhi')
```
